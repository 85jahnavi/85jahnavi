## 1.Basic Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y)
plt.xlabel('X Values')
plt.ylabel('Squared Values')
plt.title('Line Plot of y = x^2')
plt.grid(True)
plt.show()
```

## 2.Bar Plot

```python
import matplotlib.pyplot as plt

students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]

plt.bar(students, marks)
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks Scored by Students')
plt.show()
```

## 3.Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)

plt.scatter(x, y, c='red', s=100, label='Data Points')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.title('Scatter Plot of Random Points')
plt.legend()
plt.show()
```

## 4.Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(100)

plt.hist(data, bins=15, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Values')
plt.grid(True)
plt.show()
```

## 5.PieChart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D', 'E']
sizes = [20, 30, 25, 15, 10]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

plt.pie(sizes, labels=labels, colors=colors, startangle=90)
plt.title('Percentage Distribution of Categories')
plt.axis('equal')
plt.show()
```

## 6.BoxPlot(SeaBorn)

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]

sns.boxplot(data=marks, color='lightblue')
plt.title('Box Plot of Marks Scored by Students')
plt.xlabel('Marks')
plt.show()
```

## 7.Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

matrix = np.random.randint(1, 11, size=(5, 5))

sns.heatmap(matrix, annot=True, cmap='YlGnBu', cbar=True)
plt.title('Heatmap of Random 5x5 Matrix')
plt.show()
```

## 8.Comparision of matplotlib and seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
x = np.arange(1, 6)
y = [5, 9, 7, 10, 6]
data = pd.DataFrame({'X': x, 'Y': y})
plt.plot(x, y, marker='o', color='b')
plt.title('Matplotlib Line Plot')
plt.show()
sns.lineplot(x='X', y='Y', data=data, marker='o', color='r')
plt.title('Seaborn Line Plot')
plt.show()
```

## 9.Pair Plot

```python
import seaborn as sns
import pandas as pd

# Create a small dataset
data = {
    'Age': [22, 25, 30, 35, 40, 45, 50, 55, 60, 65],
    'Height': [160, 165, 170, 175, 180, 185, 190, 195, 200, 205],
    'Weight': [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]
}
df = pd.DataFrame(data)

# Create a pair plot
sns.pairplot(df)
plt.show()
```

## 10.Save and Style Plots

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.cos(x)
plt.style.use('ggplot')
plt.plot(x, y, label='Cosine Wave')
plt.title('Cosine Wave Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()
plt.savefig('cosine_wave_plot.png')
plt.savefig('cosine_wave_plot.pdf')
plt.show()
```
